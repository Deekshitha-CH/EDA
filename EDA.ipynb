DATA-SET

```python
import pandas as pd

df = pd.read_csv("C:\\Users\\Deekshitha CH\\Downloads\\retail_sales_dataset.csv")
```

DISPLAY THE TABLE

```python

print(df.head())

print("\nSummary Statistics:")
print(df.describe())
```

DATA CLEANING

```python
# Convert 'Date' to datetime
df['Date'] = pd.to_datetime(df['Date'])

# Verify conversion
print("\nUpdated Data Types:")
print(df.dtypes)
```

ANALYSIS

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Distribution of Age
plt.figure(figsize=(8, 4))
sns.histplot(df['Age'], kde=True, bins=10)
plt.title('Age Distribution')
plt.show()

# Distribution of Total Amount
plt.figure(figsize=(8, 4))
sns.histplot(df['Total Amount'], kde=True, bins=10)
plt.title('Total Amount Distribution')
plt.show()
```

CATEGORICAL DATA

```python
# Countplot for Gender
plt.figure(figsize=(6, 4))
sns.countplot(x='Gender', data=df, palette='pastel')
plt.title('Gender Distribution')
plt.show()

# Countplot for Product Category
plt.figure(figsize=(8, 4))
sns.countplot(y='Product Category', data=df, palette='muted')
plt.title('Product Category Distribution')
plt.show()
```

GENDER CATEGORY VS TOTAL AMOUNT

```python
# Boxplot for Total Amount by Gender
plt.figure(figsize=(8, 6))
sns.boxplot(x='Gender', y='Total Amount', data=df, palette='coolwarm')
plt.title('Total Amount by Gender')
plt.show()
```

PRODUCT CATEGORY VS TOTAL AMOUNT

```python
# Barplot for Total Amount by Product Category
plt.figure(figsize=(8, 6))
sns.barplot(x='Product Category', y='Total Amount', data=df, estimator=sum, palette='viridis')
plt.title('Total Amount by Product Category')
plt.show()
```

TIME BASED ANALYSIS

```python
# Monthly Sales Trend
df['Month'] = df['Date'].dt.to_period('M')  # Extract Month
monthly_sales = df.groupby('Month')['Total Amount'].sum()

# Plot Monthly Sales
plt.figure(figsize=(10, 6))
monthly_sales.plot(kind='line', marker='o')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.grid()
plt.show()
```
